### 3. Handling Missing Values

```python
mean_weight = bmdata['Item_Weight'].mean()
mean_weight
```

```python
bmdata['Item_Weight'].fillna(mean_weight, inplace = True)
bmdata.isna().sum()
```

Here we replaced the missing values present in the Item_weight category with the mean of that column

#### Next we deal with the missing values from the Outlet_Size column

```python
bmdata['Outlet_Size'].mode()
```

```python
# we create a pandas.dataframe that identifies outlet_size mode values by outlet_type
mode_of_Outlet_size = bmdata.pivot_table(values='Outlet_Size', columns ='Outlet_Type', aggfunc=(lambda x:x.mode()[0]))
```

```python
print(mode_of_Outlet_size)
```

```python
# create a list of bool output for missing values
miss_values = bmdata['Outlet_Size'].isna()
miss_values
```

```python
# we fill the missing values boolean array with .loc and apply with mode of the Outlet_Type value
bmdata.loc[miss_values,'Outlet_Size'] = bmdata.loc[miss_values, 'Outlet_Type'].apply(lambda x:mode_of_Outlet_size[x])
```

```python
bmdata.isnull().sum()
```

Now we have dealth with the missing values!

### 5. Data Preprocessing

```python
bmdata.head()
```

```python
bmdata['Item_Fat_Content'].value_counts()
```

```python
bmdata.replace({'Item_Fat_Content':{'low fat' : 'Low Fat', 'LF': 'Low Fat','reg':'Regular'}}, inplace = True)
```

```python
# aggregating data to map correctly with single labeling
bmdata['Item_Fat_Content'].value_counts()
```

### 6. Label Encoding

```python
# label encoding all categorical columns
encoder = LabelEncoder()
for i,col in enumerate(bmdata):
    if type(bmdata) == 'object':
        bmdata[str(col)]=encoder.fit_transform(bmdata[str(col)])
    else:
        continue
```

```python
bmdata.head()
```

### 7. Splitting Data into Features and Target

```python
X=bmdata.drop(columns='Item_Outlet_Sales',axis=1)
Y=bmdata['Item_Outlet_Sales']
```

```python
X.head()
```

```python
Y.head()
```

### 7. Splitting Data into Training and Testing


```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, random_state=2)
```

```python
print(X_train.shape, X_test.shape)
```

### 8. Model Training

```python
model = XGBRegressor()
model.fit(X_train, Y_train)
```

```python
# predictions
train_data_predict = model.predict(X_train)
```

```python
## RSquared Value
r2_train = metrics.r2_score(Y_train, train_data_predict)
print(r2_train)
```

```python
test_data_predict = model.predict(X_test)
```

```python
r2_test = metrics.r2_score(Y_test,test_data_predict)
print(f'And Finally! The r-squared value for the model using XGBregressor is {r2_test}!')
```
