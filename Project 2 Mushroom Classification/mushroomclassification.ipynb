## Importing the libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.svm import SVC
```

## Looking into the Dataset

```python
mush = pd.read_csv('mushrooms.csv')
mush.head()
```

## Data Preprocessing-Label Encoding

```python
mapping = list()
encoder = preprocessing.LabelEncoder()

#for every feature we will transform the values into an integer

for col in range(len(mush.columns)):
    mush[mush.columns[col]] = encoder.fit_transform(mush[mush.columns[col]])
    mapping_dict={index:label for index,label in enumerate(encoder.classes_)}
    mapping.append(mapping_dict)
```

```python
mapping
```

## Splitting the Data into X and Y

```python
X = mush.drop('class', axis=1)
Y = mush['class']
```

```python
X
```

```python
Y
```

```python
# normalizing the data
scaler= preprocessing.StandardScaler()
```

```python
X = pd.DataFrame(scaler.fit_transform(X), columns = X.columns)
X
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2)
```

## Model Selection and Score-Logistic Regression and SVM

```python
lrmodel = LogisticRegression()
svmmodel=SVC(C=1.0,kernel='rbf')
```

```python
lrmodel.fit(X_train, Y_train)
```

```python
svmmodel.fit(X_train,Y_train)
```

```python
print(f"Logistic Regression model has an accuracy score of {lrmodel.score(X_test, Y_test)}")
```

```python
print(f"Support Vector Machine {lrmodel.score(X_test, Y_test)}")
```
